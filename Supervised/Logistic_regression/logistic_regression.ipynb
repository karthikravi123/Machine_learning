```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
##create dataset

#X,y= make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)

X,y= make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=15)
#3x= indpedent
##y-dependent
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train.shape
```

```python
##implement logistic regression
##Model Training
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred = logistic.predict(X_test)
print(y_pred)
```

```python
# logistic.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = accuracy_score(y_test,y_pred)
print(score)

print("---------------------")
cm = confusion_matrix(y_test,y_pred)
print(cm)

print("---------------------")
print(classification_report(y_test,y_pred))

##classsification -precision ,recall
```

Hyperparameter tunning - model selects best attributes,attributes used to train over model so that is produce good accuracy and doesnt do overfitting

Hyper parameter tunning and cross validation

-Grid Search hyper parameter

```python
model = LogisticRegression()

##reffered from documentation
##sklearn logisticregression
penalty = ['l1','l2','elasticnet']
c_values = [100,10,1.0,0.1,0.01]
solver = ['newton-cg','lbfgs','liblinear','sag','saga']
```

```python
dict(penalty=penalty,C=c_values,solver=solver)
```

```python
# params = dict(penalty=penalty,C=c_values,solver=solver)

##select the best combination
```

```python
params = [
    {'penalty': ['l2'], 'solver': ['newton-cg', 'lbfgs', 'sag'], 'C': c_values},
    {'penalty': ['l1'], 'solver': ['liblinear', 'saga'], 'C': c_values},
    {'penalty': ['elasticnet'], 'solver': ['saga'], 'C': c_values, 'l1_ratio': [0.5, 0.7, 0.9]}
]
```

```python
##gridsearch cv
##find out best parameter best fit for this dataset



from sklearn.model_selection import GridSearchCV,StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_test,y_pred)
print(score)

print("---------------------")
cm = confusion_matrix(y_test,y_pred)
print(cm)

print("---------------------")
print(classification_report(y_test,y_pred))

##classsification -precision ,recall
```

One more Hyperparameter techinques -
 Randomised tunning -it pick up random parameter ,rando, attribute and the it will find out which parameter is suitable


 large dataset - take less amount of time (advantage)

```python
model = LogisticRegression()
randomcv =RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```
