```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
##create instance of One hot encoder

encoder = OneHotEncoder()
```

```python
##perform fit and transform
encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Load the dataset
df = sns.load_dataset('tips')

# Select categorical columns to encode
categorical_cols = ['sex', 'smoker', 'day', 'time']

# Initialize the encoder
encoder = OneHotEncoder(sparse_output=False)  # sparse=False in older versions

# Fit and transform
encoded_array = encoder.fit_transform(df[categorical_cols])

# Optional: convert to DataFrame for readability
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categorical_cols))

# Combine with original data (if needed)
df_combined = pd.concat([df.drop(columns=categorical_cols), encoded_df], axis=1)

# Display result
print(df_combined.head())
```

```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Load the dataset
df = sns.load_dataset('tips')

# Select categorical columns to encode
categorical_cols = ['sex', 'smoker', 'day', 'time']

# Initialize the encoder
encoder = OneHotEncoder(sparse_output=False)  # sparse=False in older versions

# Fit and transform
encoded_array = encoder.fit_transform(df[categorical_cols])

# Optional: convert to DataFrame for readability
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out())

# Combine with original data (if needed)
df_combined = pd.concat([df.drop(columns=categorical_cols), encoded_df], axis=1)

# Display result
print(df_combined.head())
```
