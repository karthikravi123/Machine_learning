```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
iris = load_iris()
```

```python
pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'])
```

```python
##independent features
X= pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'])
```

```python
##dependent feature
y = iris['target']
```

```python
##train test

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=10)
```

```python
## no need standrizartion and normalization in Decision treee
##post prunning is done for smaller dataset
```

```python
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()

#treeclassifier = DecisionTreeClassifier(max_depth=2)
treeclassifier.fit(X_train,y_train)
```

```python
##Visualize the Decision tree 
from sklearn import tree
plt.figure(figsize=(20,20))
tree.plot_tree(treeclassifier,filled=True)
```

```python
y_pred =treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
